```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.integrate import ode


def lorenz_63(t, coord):
    x, y, z = coord
    return [10 * (y - x),
            x * (28 - z) - y,
            x * y - (8 / 3) * z]


def evaluate_lorenz(initial_point=None, dt=0.02, N=4000):
    if initial_point is None:
        initial_point = 1 * (2 * np.random.rand(3) - 1)

    r = ode(lorenz_63).set_integrator('vode', method='bdf')
    r.set_initial_value(initial_point, 0)
    sol = [initial_point, ]
    while len(sol) <= N:
        sol.append(r.integrate(r.t+dt))
    return np.array(sol).T
```

```python
data = evaluate_lorenz([1, 1, 1], dt=0.01, N=4000)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(*data)
plt.show()
```

```python
from numpy.linalg import cond, norm, inv


w = 0.05
b = 4
D_r = 1000
beta_val = [10**(n/2) for n in range(-30, 0)]

U_o = data[:, 1:]
U_i = data[:, :-1]

# without coefficient
Tau = None
for _ in range(10):
    tau = np.zeros([30, 1])
    for i, beta in enumerate(beta_val):
        W_in = w * (2 * np.random.rand(D_r, 3) - 1)
        b_in = b * (2 * np.random.rand(D_r, 1) - 1)

        Phi = np.tanh(W_in @ U_i + b_in)
        W_lr = U_o @ Phi.T @ inv(Phi @ Phi.T + beta * np.identity(D_r))
        
        b_in = b_in.reshape([D_r, ])
        u = data[:, 0]
        u = W_lr @ np.tanh(W_in @ u + b_in)

        steps = 0
        e = (np.dot(data[:, steps+1] - u, data[:, steps+1] - u) / np.dot(data[:, steps+1], data[:, steps+1]))
        while e <= 0.05:
            steps += 1
            u = W_lr @ np.tanh(W_in @ u + b_in)
            e = (np.dot(data[:, steps+1] - u, data[:, steps+1] - u) / np.dot(data[:, steps+1], data[:, steps+1]))
        
        tau[i] = steps * 0.02
        
    if Tau is None:
        Tau = tau
    else:
        Tau = np.concatenate((Tau, tau), axis=1)
        
M = np.apply_along_axis(norm, 1, Tau)


# with coefficient
Tau = None
for _ in range(10):
    tau = np.zeros([30, 1])
    for i, beta in enumerate(beta_val):
        W_in = w * (2 * np.random.rand(D_r, 3) - 1)
        b_in = b * (2 * np.random.rand(D_r, 1) - 1)

        Phi = (1/np.sqrt(D_r)) * np.tanh(W_in @ U_i + b_in) # here
        W_lr = U_o @ Phi.T @ inv(Phi @ Phi.T + beta * np.identity(D_r))
        
        b_in = b_in.reshape([D_r, ])
        u = data[:, 0]
        u = W_lr @ ((1/np.sqrt(D_r)) *  np.tanh(W_in @ u + b_in)) # here

        steps = 0
        e = (np.dot(data[:, steps+1] - u, data[:, steps+1] - u) / np.dot(data[:, steps+1], data[:, steps+1]))
        while e <= 0.05:
            steps += 1
            u = W_lr @ ((1/np.sqrt(D_r)) *  np.tanh(W_in @ u + b_in)) # here
            e = (np.dot(data[:, steps+1] - u, data[:, steps+1] - u) / np.dot(data[:, steps+1], data[:, steps+1]))
        
        tau[i] = steps * 0.02
        
    if Tau is None:
        Tau = tau
    else:
        Tau = np.concatenate((Tau, tau), axis=1)
        
M_c = np.apply_along_axis(norm, 1, Tau)

plt.plot(beta_val, M_c, label='with coefficient')
plt.plot(beta_val, M, label='without coefficient')
plt.xscale('log')
plt.legend()

plt.show()
```

```python
from numpy.linalg import cond, norm, inv


w = 0.05
b = 4
D_r = 1000
beta_val = [10**n for n in range(-15, 1)]

U_o = data[:, 1:]
U_i = data[:, :-1]

E = None
C = None
for _ in range(50):
    condi = np.empty([16, 1])
    error = np.empty([16, 1])
    for i, beta in enumerate(beta_val):
        W_in = w * (2 * np.random.rand(D_r, 3) - 1)
        b_in = b * (2 * np.random.rand(D_r, 1) - 1)

        Phi = np.tanh(W_in @ U_i + b_in)
        P = Phi @ Phi.T + beta * np.identity(D_r)
        condi[i] = cond(P)
        error[i] = norm(inv(P) @ P - np.identity(D_r))
    
    if C is None:
        C = condi
    else:
        C = np.concatenate((C, condi), axis=1)
        
    if E is None:
        E = error
    else:
        E = np.concatenate((E, error), axis=1)        
        
E = np.apply_along_axis(np.mean, 1, E)
C = np.apply_along_axis(np.mean, 1, C)
        
fig, ax = plt.subplots(1, 2)
ax[0].plot(beta_val, error, '.')
ax[0].set_xscale('log') 
ax[0].set_yscale('log')
ax[0].set_xlabel('beta')
ax[0].set_ylabel('error of PP^-1')

ax[1].plot(beta_val, condi, '.')
ax[1].set_xscale('log') 
ax[1].set_yscale('log')
ax[1].set_xlabel('beta')
ax[1].set_ylabel('condition number')

fig.suptitle('Average over 50 realizations.')
plt.show()
```
